```python
df = pd.read_csv("microloan.csv")
#Printing the head
df.head()
```

```python
df = df.dropna()
df
```

```python
df.drop(['Loan Date','Block','Street Direction','Street Name','Street Type','City','Zip','Ward','Latitude','Longitude','Location'], axis=1, inplace=True)
```

```python
df = df.sort_values(['Amount'],ascending=False)
```

```python
df1 = pd.get_dummies(df.Lender, prefix='Lender')
df = df.join(df1)
df.drop(['Lender'], axis=1, inplace=True)
df2 = pd.get_dummies(df.Industry, prefix='Industry')
df = df.join(df2)
df.drop(['Industry'], axis=1, inplace=True)
```

```python
from sklearn.metrics import jaccard_score
J1 = (jaccard_score(df['Lender_Accion'], df['Industry_Arts/Crafts Production']))
J2 = (jaccard_score(df['Lender_CNI'], df['Industry_Arts/Crafts Production']))
J3 = (jaccard_score(df['Lender_WBDC'], df['Industry_Arts/Crafts Production']))
print(J1)
print(J2)
print(J3)
if max(J1,J2,J3) == J1:
  print('ACCION')
elif max(J1,J2,J3) == J2:
  print('CNI')
elif max(J1,J2,J3) == J3:
  print('WBDC')
```
